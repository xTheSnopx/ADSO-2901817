1.   Revisión y tratamiento de los datos NaN

```python
df.isnull().sum()
```

```python
df['Age'].isnull().sum()
```

```python
mediana_edad = df['Age'].median()
print("Mediana de Age:", mediana_edad)
```

```python
df['Age'].fillna(mediana_edad, inplace=True)
```

```python
df['Age'] = df['Age'].fillna(mediana_edad)
```

```python
df['Age'].isnull().sum()
```

```python
df['Gender'].value_counts()
```

```python
moda_genero = df['Gender'].mode()[0]
print(f"La moda es: {moda_genero}")
```

```python
df['Gender'] = df['Gender'].fillna(moda_genero)
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].str.lower()
```

```python
df['Gender'] = df['Gender'].replace({
    'femenino': 'Femenino',
    'femenino': 'Femenino',
    'masculino': 'Masculino',
    'masculinoooo': 'Masculino',
    'masculino.....//': 'Masculino',
    'femenino': 'Femenino'
})
```

```python
df['Gender'].value_counts()
```

```python
import re
df['Gender'] = df['Gender'].apply(lambda x: re.sub(r'[^a-zA-Z]', '', str(x)))
```
